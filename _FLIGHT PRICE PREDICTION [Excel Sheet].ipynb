```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By

import time
```

```python
driver= webdriver.Chrome(r'C:\Users\user\Downloads\chromedriver_win32 (2)\chromedriver.exe')
time.sleep(2)
```

```python
url='https://flight.yatra.com/air-search-ui/dom2/trigger?type=O&viewName=normal&flexi=0&noOfSegments=1&origin=DEL&originCountry=IN&destination=BOM&destinationCountry=IN&flight_depart_date=26%2F09%2F2022&ADT=1&CHD=0&INF=0&class=Economy&source=fresco-home&unqvaldesktop=1050749650766'
driver.get(url)
```

```python
air_detail=[]
flight_d=driver.find_elements(By.XPATH,"//div[@class='flight-det table full-width clearfix']")
for i in flight_d:
    air_detail.append(i.text)
len(air_detail)    
```

```python
air_detail
```

```python
airline=[]
flight_code=[]
departure_time=[]
source_city=[]
arrival_time=[]
destination_city=[]
duration=[]
total_stops=[]
target_variable_price=[]
time.sleep(2)
```

```python
time.sleep(4)
airline_n=driver.find_elements(By.XPATH,"//div[@class='flight-det table full-width clearfix']")
for i in airline_n[0:1500]:
    airline.append(i.text.split('\n')[0])
    
flight_c=driver.find_elements(By.XPATH,"//div[@class='flight-det table full-width clearfix']")
for j in flight_c[0:1500]:
    flight_code.append(i.text.split('\n')[1])
    
dept_time=driver.find_elements(By.XPATH,"//div[@class='flight-det table full-width clearfix']")
for m in dept_time[0:1500]:
    departure_time.append(i.text.split('\n')[2])    

src_city=driver.find_elements(By.XPATH,"//div[@class='flight-det table full-width clearfix']") 
for k in src_city[0:1500]:
    source_city.append(i.text.split('\n')[3])
    
arl_time=driver.find_elements(By.XPATH,"//div[@class='flight-det table full-width clearfix']")
for n in arl_time[0:1500]:
    arrival_time.append(i.text.split('\n')[4])    
    
destn=driver.find_elements(By.XPATH,"//div[@class='flight-det table full-width clearfix']")    
for l in destn[0:1500]:
    destination_city.append(i.text.split('\n')[5])  

durn=driver.find_elements(By.XPATH,"//div[@class='flight-det table full-width clearfix']") 
for o in durn[0:1500]:
    duration.append(i.text.split('\n')[6])
    
tot_stops=driver.find_elements(By.XPATH,"//div[@class='flight-det table full-width clearfix']")    
for p in tot_stops[0:1500]:
    total_stops.append(i.text.split('\n')[7])

price=driver.find_elements(By.XPATH,"//div[@class='flight-det table full-width clearfix']")
for q in price[0:1500]:
    target_variable_price.append(i.text.split('\n')[8])
```

```python
len(airline),len(flight_code),len(departure_time),len(source_city),len(arrival_time),len(destination_city),len(duration),len(total_stops),len(target_variable_price)
```

```python
df=pd.DataFrame({'Airline Company':airline,'Flight Code':flight_code,'Departure Time':departure_time,'Source City':source_city,'Arrival Time':arrival_time,'Destination City':destination_city,'Duration':duration,'Total Stops':total_stops,'Price':target_variable_price})
df
```
